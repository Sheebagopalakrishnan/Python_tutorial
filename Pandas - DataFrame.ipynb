```python
# 1. Using dict
d = {'a':0, 'b':1,'c':2}
p = pd.DataFrame(d, index=[1,2])       #must specify index
print(p)

# 2. Using dict of series

d1 = {'col1':pd.Series([1,2,3,4], index = ['row1','row2','row3','row4']), 'col2':pd.Series([10,20,30],index=['row1','row2','row3'])}
p1 = pd.DataFrame(d1)
print(p1)

# 3. Using dict of Array/Lists

d2 = {'Name': ['she','vig','dad','mom','muki'], 'Age': [26,30,45,55,22] }      #takes the default index
p2 = pd.DataFrame(d2)
print(p2)

p3 = pd.DataFrame(d2, index= ['row1','row2','row3','row4','row5'])             #Or you can specify the index values
print(p3)

# 4. Using list of dicts

l = [{'col1':10, 'col2':20, 'col3':30},{'col1':11,'col2':22,'col3':33,'col4':44}]
p4 = pd.DataFrame(l)
print(p4)

p5 = pd.DataFrame(l, columns = ['col1','col2'], index = [1,2])                 #Can also filter out the cols you need
print(p5)
```

```python
#Basic DF Selection, Addition & Deletion 

df = pd.DataFrame({'one': [1,2,3,4,5,6],'two':[1,2,3,4,5,6]})
print(df)

df['one']                                     #prints column 'one'
df['three'] = df['one'] + df['two']           #creates new column 'three'
df['flag1'] = df['one'] > 2
df['flag2'] = df['two'] > 3
df['one_trunc'] =df['one'][:2]
print(df)

del df['flag2']                               #Delete a column using 'def' or 'pop' keyword
print(df)
df.pop('flag1')                  
print(df)

df['four'] = 'copy'                           #scalar value gets filled in the column four
print(df)

df.insert(1,'temp',df['four'])                #inserts column at specific location

print(df)

df.assign(temp2 = df['one'] / df['two'])      #allows to create new columns that are potentially derived from existing columns

df.assign(col1 = lambda x : x['one']+x['two'], col2 = lambda x : x['one'] * x['two'])
```

```python
print(df.head())                                  #By default, prints first five rows
print(df.head(2))
print(df.tail())                                  #prints the last five rows
print(df.columns)
print(df.index)
print(df.dtypes)
print(df.describe())                              #statistical summary
print(df.T)
print(df.sort_index(axis = 1,ascending =False))   #sorts the axis (Rearrange row/column to ASC/DESC using axis=0/1)
print(df.sort_values(by ='one',ascending =False)) #sorts by values 
dff = df.copy()
dff['one'].mean()
```

```python
#Indexing and Slicing

df1 = pd.DataFrame({'one': [1,2,3],'two':[1,2,3]}, index = ['a','b','c'])
print(df1)

#Selection by label
print(df1.loc['c'])                   #Select row by index_labelname/row-label
print(df1.loc['a':'b', ['one']])      #retruns label a to b and selects only 'one' column
print(df1['one']['a':'b'])            #equivalent to previous one
 
#Selection by position
print(df1.iloc[1])                    #Select row by integer location (rownumber)
print(df1[1:2])                       #filters row starting from 1 and ends with 3-1, equivalent to above

#Boolean indexing
print(df1[df1['one'] > 2])                 
print(df1[df1.one > 2])                      #equivalent to above
print(df1[df1>0])

#filter using isin()
df1['test'] = ['T1','T2','T3']
print(df1[df1['test'].isin(['T2','T3'])])   #filter by values in column

df1[df1['one'] > 2] = 4
print(df1)

df1.info()
```

```python
columns = ['name', 'age', 'gender', 'job']
user1 = pd.DataFrame([['alice', 19, "F", "student"],['john', 26, "M", "student"]],columns=columns)
user2 = pd.DataFrame([['eric', 22, "M", "student"],['paul', 58, "F", "manager"]],columns=columns)
user3 = pd.DataFrame(dict(name=['peter', 'julie'],age=[33, 44], gender=['M', 'F'],job=['engineer', 'scientist']))

#concat all user
users = pd.concat([user1,user2,user3])
print(users)
user= user1.append(user2)
print(user)

#Join
user4 = pd.DataFrame(dict(name=['alice', 'john', 'eric', 'julie'],height=[165, 180, 175, 171]))
inner_join = pd.merge(users,user4,on="name")
print(inner_join)

outer_join=pd.merge(users,user4,on="name",how="outer")
print(outer_join)

#Group by data
print(users.groupby("gender").count())      #SPECIFY AGGREAGTE FUNCTION
print(users.groupby("job").age.mean())
```

```python
#Row Selection - Basics

print(users.head(1))
print(users[:1])
print(users.loc[0])     # Returns all the rows which has index as 0
print(users.iloc[0])    #Returns the first row
print(users.iloc[0,0])
  

print(users.tail(1))
print(users[-1:])
print(users.iloc[-1])


print(users[::-1])      #(Reverse)

#Row Selection - Filtering

print(users[users['age']>25])                            #Filter row [imagine as Select * from users where age > 25]
print(users[users['age']>25]['name'])                    #FIlter rows for single col
print(users[users['age']>25].name)
print(users[users['age']>25][['name','gender']])         #Filters rows for multiple cols
```

```python
#Duplicate data removal

user5 = users.append(user1)          #Just to add duplicates

user5.describe(include="all")        #To check the total rows and unique rows

#Entire table check
print(user5.duplicated())
print(user5.duplicated().sum())      #Count the no of duplicates
print(user5[user5.duplicated()])     #Displays duplicate rows

#Row-wise check
print(user5['age'].duplicated())     #Column-wise duplicates sum
print(user5.age.duplicated())
print(user5[["age","name"]].duplicated())
print(user5.duplicated(["age","name"]))
print(user5.duplicated(["age","name"]).sum())

#Delete duplicates
user6=user5.drop_duplicates()
print(user6)
```

```python
#Missing Data
user7 = outer_join
print(user7)

print(user7.isnull())
print(user7['height'].isnull())
print(user7['height'].isnull().sum())

print(user7[user7['height'].notnull()])   # Returns the notnull rows

#Delete missing data
user8 = user7.dropna()
print(user8)

#Fill the missing values
user9 = user7['height'].fillna(value = 100)  #saves just the column value
print(user9)

user10 = user7.copy()
user10.loc[user10['height'].isnull(),'height'] = user10['height'].mean()  #updates the table
print(user10)
```
